```python
import pandas as pd
import numpy as np
from pathlib import Path
import os
```

```python
bike_data_1 = pd.read_csv('resources/JC-202406-citibike-tripdata.csv')
bike_data_2 = pd.read_csv('resources/JC-202407-citibike-tripdata.csv')
bike_data_3 = pd.read_csv('resources/JC-202408-citibike-tripdata.csv')

```

```python
appended_data = pd.concat([bike_data_1, bike_data_2, bike_data_3], ignore_index=True)
appended_data.head()
```

```python
count_row = appended_data['ride_id'].count()
count_row
```

```python
bike_data_df = pd.DataFrame(appended_data)
```

```python
# Assuming bike_data_df is your DataFrame and 'started_at' is the column you want to convert
# Specify the format of your datetime strings including milliseconds
date_format_with_ms = '%Y-%m-%d %H:%M:%S.%f'
date_format_without_ms = '%Y-%m-%d %H:%M:%S'

bike_data_df['started_at'] = pd.to_datetime(bike_data_df['started_at'], format=date_format_with_ms, errors='coerce')

bike_data_df['started_at'] = bike_data_df['started_at'].fillna(pd.to_datetime(bike_data_df['started_at'], format=date_format_without_ms, errors='coerce'))

bike_data_df['ended_at'] = pd.to_datetime(bike_data_df['started_at'], format=date_format_with_ms, errors='coerce')

bike_data_df['ended_at'] = bike_data_df['ended_at'].fillna(pd.to_datetime(bike_data_df['ended_at'], format=date_format_without_ms, errors='coerce'))

bike_data_df['ended_at_date'] = bike_data_df['ended_at'].dt.date

filtered_data = bike_data_df[bike_data_df['started_at'].dt.month.isin([6, 7, 8])]

sampled_data = filtered_data.sample(n=3000, random_state=42)
```

```python
sampled_data.head()
```

```python
resources_folder = 'resources'
file_path = os.path.join(resources_folder, 'sampled_data.csv')
sampled_data.to_csv(file_path, index=False)
```
